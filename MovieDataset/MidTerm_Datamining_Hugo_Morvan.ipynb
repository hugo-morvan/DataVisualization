Count of release per year:

```python
count_per_year = df.groupby(['year_released'])['year_released'].count().reset_index(name='count').sort_values(['year_released'], ascending=False)

print(count_per_year.to_markdown(index=False))
```

List the name of 5 directors with least shows count. Show count includes the "type". You need to find the name of at least five directors and display the output with the director name, type, country and count.

```python
count_per_director = df.groupby(['director'])['director'].count().sort_values()
count_per_director = count_per_director.head(5)

dirs = count_per_director.index

countries = [(df.loc[df['director'] == dir]['country'].iloc[0]) for dir in dirs]
types = [(df.loc[df['director'] == dir]['type'].iloc[0]) for dir in dirs]

table = {'director': [dir for dir in dirs], 
         'type':[t for t in types], 
         'country': [c for c in countries], 
         'show-count':[count for count in count_per_director]}

tab = pd.DataFrame(table)
display(tab)
```

Pie chart of the first five country counts (including "Not Given")

```python
count_per_country = df.groupby(['country'])['country'].count().sort_values(ascending=False).head(5)
#print(count_per_country) 
plot = count_per_country.plot.pie(figsize=(5,5))
```

Pie chart of the first fice country counts (excluding "Not Given")

```python
count_per_country2=df.groupby(['country'])['country'].count().sort_values(ascending=False)
count_per_country2.drop(index="Not Given",  inplace=True)
count_per_country2 = count_per_country2.head(5)
#print(count_per_country2)
plot = count_per_country2.plot.pie(figsize=(5,5))
```

Line plot for "Type" with respect to "year_released_count" and year released.

```python
count_per_year_per_type = df.groupby(['year_released','type']).size().unstack(level=1)
count_per_year_per_type = count_per_year_per_type.fillna(0)
#print(count_per_year_per_type)
movie = count_per_year_per_type[['Movie']].copy()
show = count_per_year_per_type[['TV Show']].copy()
#print(movie)
x = movie.index
#print(x)
plt.plot(x, movie, label='Movie')
plt.plot(x, show, label='TV Show')

plt.xlabel('Year')
plt.ylabel('Count of releases')
plt.title('Count of releases of Movies and TV Shows per year')
plt.legend()

```
